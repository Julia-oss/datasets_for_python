### 1) Загрузка данных

Ссылка для скачивания файла 'contries.csv'

```python
# Импортируем библиотеки pandas и openpyxl
import pandas as pd
import openpyxl

# Читаем данные из файла
df = pd.read_csv('countries.csv')
df
```

### 2) Анализ данных

```python
# Создаем сводную таблицу
tab = df.pivot_table(index='country', columns='year', aggfunc='sum')
tab
```

### 3) Сохранение отчета

```python
# Экспортируем данные в Excel
tab.to_excel('report_2025.xlsx', sheet_name='Report', startrow=4)
```
