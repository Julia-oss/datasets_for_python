## Задание 1: Приветствие

## Задание 2: Сложение двух чисел
### Задание:

## Задание 3: Умножение чисел
### Задание:

## Задание 4: Вывод текста
### Задание:

## Задание 5: Деление чисел
### Задание:

## Задание 6: Вывод результата сложения
### Задание:

Рассчитайте их сумму и выведите результат с текстом "Результат сложения: [сумма]".

## Задание 7: Время и расстояние
### Задание:

Создайте переменную speed, которая будет хранить скорость автомобиля (в километрах в час).
Создайте переменную time, которая будет хранить время в пути (в часах).
Вычислите расстояние, которое преодолеет автомобиль, используя формулу:

distance=speed×time
Выведите результат.

## Задание 8: Разница между числами
### Задание:
